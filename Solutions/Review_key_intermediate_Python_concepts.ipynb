## For loops and list comprehensions

````python
results = []
for book in ['Harry Potter', 'War and Peace', 'Power of Habit']:
    if 'Harry' not in book:
        results.append(book)    
````

Let's do it in a list comprehension way.

````python
BOOKS = ['Harry Potter', 'War and Peace', 'Power of Habit']
results = [book for book in BOOKS if 'Harry' not in book]
````

### Exercise
Create a function in which you can pass in an arbitrary function and it will apply that to the numbers that are divisible by `divisor` between  `start` and `end`.

For example,

```
def cube(num):
    return num ** 3

apply_func_to_divisble_by(1, 20, 2, cube)
>>> [8, 64, ..., 8000]
```

Note: 8 is 2^3, 64 is 4^3, ...

````python
def apply_func_to_divisible_by(start, end, divisor, func):
    return [func(i) for i in range(start, end) if i % divisor == 0]
````

````python
apply_func_to_divisible_by(1, 20, 2, lambda val: val ** 3)
````

## `*args` and `**kwargs`

````python
def add_up(numbers):
    count = 0
    for num in numbers:
        count+= num
    return count

add_up([2, 5, 7, 8])
````

````python
def add_up_with_args(*numbers):
    count = 0
    for num in numbers:
        count+= num
    return count
````

````python
add_up_with_args(1, 2, 7)
````

````python
add_up_with_args(*[2, 28, 35, -7, 103])
````

See how you can call the below function.

````python
def get_odds(**kwargs):
    start = kwargs.get('start', 1)
    end = kwargs.get('end', 1000)
    nums = range(start, end)
    divisor = kwargs.get('divisor', 2)
    return [n for n in nums if n % divisor == 1]
````

### Application of `*args` and `**kwargs`

````python
class Animal():
    def __init__(self, name, genus, species, alive=True):
        self.name = name
        self.genus = genus
        self.species = species
        self.alive = alive
````

````python
class Person(Animal):
    def __init__(self, job, *args, health_level='average', **kwargs):
        self.job = job
        self.health_level = health_level
        super().__init__(*args, **kwargs)
````

````python
Person('engineer', 'bob', 'homo', 'sapiens')
````

## Operator overloading

Let's create a class called `DataFrame`.

It should take `data` which is a dictionary where the keys are column names and the values are a list of values for that column.

````python
data = {"age": [50, 25, 79], "name": ["Harry", "Suzy", "Alberto"], "location": ["New York", "San Jose", "Virginia"]}
````

````python
class DataFrame():
    def __init__(self, data):
        self.data = data

    def __repr__(self):
        return "DataFrame with columns {}".format(','.join(self.data.keys()))
    
    def __getitem__(self, key):
        if isinstance(key, str):
            return self.data[key]
        else:
            return {col: self.data[col] for col in key}
````

````python
df = DataFrame({"age": [50, 25, 79], "name": ["Harry", "Suzy", "Alberto"], "location": ["New York", "San Jose", "Virginia"]})
````

````python
df["age"]
````
